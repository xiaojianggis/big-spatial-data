```python
import geopandas as gpd
from geopandas import GeoDataFrame
import os, os.path
import pandas as pd
from shapely.geometry import Point
```

#### 1. Read csv file and output as shapefile
Let suppose we have a csv file (or other spreadsheets) with attributes and coordindates. We want to convert the csv file into a shapefile in order for further spatial analysis. The first part is to read the csv file and write the result to a shapefile. You will need modules of shapely (for geometric operations), pandas (reading csv file), and geopandas (write to shapefile). 

```python
# read the csv file and create shapefile
coffee_shop_csv = 'sample-data/cambridge_coffee_shops.csv'
coffee_shop_sheet = pd.read_csv(coffee_shop_csv)
coffee_shop_sheet.head()
```

```python
coffee_shop_sheet['lat'].head()
coffee_shop_sheet['lon'].head()
```

```python
geometry = [Point(xy) for xy in zip(coffee_shop_sheet['lon'], coffee_shop_sheet['lat'])]
crs = {'init': 'epsg:4326'} #http://www.spatialreference.org/ref/epsg/4326/
geo_df = GeoDataFrame(coffee_shop_sheet, crs=crs, geometry=geometry)

geo_df.to_file(driver='ESRI Shapefile', filename='sample-data/cambridge_coffee_shops.shp')
```

#### 2. Read shapefile using GeoPandas
Before do real spatial analyses, you need to read shapefile and access the attributes of the shapefile. It is very easy to read shapefile in GeoPandas. Using the function of "read_file", you can read the shapefile. The function will return a dataframe with geometry column. 

```python
cof_df = gpd.read_file('sample-data/Cambridge-coffee-shop.shp') #coffee shop shapefile
nb_df = gpd.read_file('sample-data/cambridge_neighborhoods.shp') #neighborhood shapefile
cof_df.head() #print the first five records
```

#### 3. Plot the location of these coffee shops
Sometimes, we want to know how this coffee shops distributes spatially. Here we can map and visualize this distribution without using GIS software. The geopandas dataframe can be ploted directly.

```python
ax = cof_df.plot(color="red", figsize=(8,8))
# ax.set(xticks=[], yticks=[]) #with no xlabel and ylabel
```

You can also map the coffee shop and using different color to represent coffee shopes in different zip code area

```python
cof_df.plot(column='zip', figsize=(8,8))
```

#### Plot these two layers together
We like to have a basemap on the coffee shop map. We can use the neighborhood layer as the basemap.

```python
import matplotlib.pyplot as plt

base=nb_df.plot(color='white', edgecolor='black', figsize=(8,8))
cof_ax = cof_df.plot(ax=base, marker='o', column='zip', figsize=(8,8))
cof_ax.set(xticks=[], yticks=[]) # remove the xlabel and ylabel
```

Using the matplotlib to plot the figures

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
nb_df.plot(ax=ax, color='white', edgecolor='black', figsize=(8,8))
cof_df.plot(ax=ax, column='zip', marker='o', markersize=8)
# ax.set(xticks=[], yticks=[])
```

#### 4. Managing projections of the shapefile
Projects matter in spatial analysis. You need to be careful about map project before doing spatial analysis. In this section, you will know how to read the projection of shapefile, and reproject shapefiles.

You can access to the projection information and manage map projections easily using GeoPandas.

```python
# print the coordinate reference system (crs)
print (cof_df.crs)
print (nb_df.crs)
```

#### change map projection
We all know how important is the projection. For example, you want to do the buffer analysis. However, the original cof layer is in projection of WGS84. Then you want to use a local projection with unit of feet. Then you may consider the Massachusetts land projection http://spatialreference.org/ref/epsg/?search=massachusetts&srtext=Search. 

```python
# the crs of Massachusetts mainland
mass_crs = {'init': u'epsg:2249'}

ax = nb_df.plot()
ax.set_title("WGS84 (lat/lon)");
```

#### reproject to the local projection of Massachusetts

```python
fig, ax3=plt.subplots()

nb_df_mass = nb_df.to_crs(mass_crs) # world.to_crs(epsg=3395) would also work
cof_df_mass = cof_df.to_crs(mass_crs)

nb_df_mass.plot(ax=ax3, color='white', edgecolor='black')
cof_df_mass.plot(ax=ax3, color='red', markersize=3)
ax3.set_title("Massachusetts projection")
ax3.set(xticks=[], yticks=[])
```

#### 5. Geometric manipulation using geopandas
Just like I mentioned previously, the geopandas wraps fiona, shapely, and pandas. The geometric manipulation in geopandas is based on the shapely libarary. For more details about the module shapely, check http://shapely.readthedocs.io/en/stable/manual.html 

Let take look at the geometry column of the geopandas data frame.


```python
nb_df_mass['geometry']
```

each record in the dataframe is a shapely geometrical object

```python
nb_df_mass['geometry'][0]
```

```python
nb_df_mass['geometry'][1]
```

Based on this, we can do the buffer analysis on the coffee shops

```python

fig, ax4=plt.subplots()

cof_df_mass['geometry'] = cof_df_mass.buffer(150)
nb_df_mass.plot(ax=ax4)
cof_df_mass.plot(ax=ax4, color='red')
```

Overlay of the point and polygon features

```python
nb_inter = gpd.overlay(nb_df_mass, cof_df_mass, how='intersection')
nb_inter.plot(edgecolor='k', color='cyan')
```

After intersection, the intersected features will have a merged attributes

```python
nb_inter
```
